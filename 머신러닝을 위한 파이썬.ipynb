### 리스트 메소드(함수)(List Methods)

▣ 리스트 끝에 요소 추가하기(append)
* 리스트에 요소를 추가하는 함수
* append(x)는 특정 리스트의 **끝에** x라는 요소를 추가하는 함수

▣ 리스트의 원하는 위치에 요소 삽입하기(insert)
* insert(a,b)로 사용 : **a위치에 b요소를 삽입**한다는 의미

▣ 리스트 정렬하기(sort)
* sort함수는 숫자, 알파벳 등을 오름차순으로 정렬

▣ 리스트 뒤집기(reverse)
* 특정 기준(오름차순 또는 내림차순)으로 정렬하는 것이 아니라 현재 있는 요소들을 단지 거꾸로 뒤집는 함수

▣ 요소의 위치 반환(index)
* index(x)는 리스트에 x라는 요소가 있을 때, 어떤 위치, 어떤 index에 있는지 알려줌
* 리스트에 x가 없을 때는 오류를 반환

▣ 리스트 요소 제거하기(remove)
* 리스트에서 나오는 첫번째 특정 요소를 삭제하는 함수
* remove(x)는 리스트에서 첫번째로 등장하는 **x라는 요소를 삭제**하는 것
* 요소가 리스트에 존재하지 않는다면 오류를 반환

▣ 리스트 요소 꺼내기(pop)
* pop()는 마지막 요소
* pop(x) 와 같이 pop함수 뒤에 하나의 인자를 넣어주면 **x위치에 있는 요소를 꺼냄**

▣ 리스트의 요소 개수 세기(count)
* count(x) : 리스트에서 x라는 요소의 개수를 반환

▣ 리스트 확장하기(extend)
* extend함수는 원래의 리스트에 리스트를 더하는 함수
